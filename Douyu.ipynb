# 某一时间点斗鱼颜值主播的地域分布图

```python
import pymongo
from string import punctuation
import charts
```

```python
client = pymongo.MongoClient('localhost', 27017)
Douyu = client['Douyu']
item_info = Douyu['images']
```

```python
for i in item_info.find().limit(30):
    print(i['anchor_city'])
```

```python
area_list=[]
for i in item_info.find():
    area_list.append(i['anchor_city'])
area_index = list(set(area_list))
print(area_index)
```

```python
post_times = []
for index in area_index:
    post_times.append(area_list.count(index))
print(post_times)
```

```python
def data_gen(types):
    length = 0
    if length <= len(area_index):
        for area,times in zip(area_index,post_times):
            data = {
                'name':area,
                'data':[times],
                'type':types
            }
            yield data
            length +=1
```

```python
for i in data_gen('column'):
    print(i)
```

```python
series = [data for data in data_gen('column')]
charts.plot(series, show='inline', options=dict(title=dict(text='某一时间点斗鱼颜值主播的地域分布柱状图')))
```

```python
def data_gen_pie():
    length = 0
    if length <= len(area_index):
        for area,times in zip(area_index,post_times):
            data = [ area,times]
            yield data
            length +=1
```

```python
series =  [{
    'type': 'pie',
    'name': 'Browser share',
    'data': [data for data in data_gen_pie()]
}]

charts.plot(series, options={'title': {'text': '某一时间点斗鱼颜值主播的地域分布饼图'}}, show='inline')
```

```python
pipeline = [
    {'$group':{'_id':'$anchor_city','counts':{'$sum':1}}},
    {'$sort' :{'counts':-1}},
    {'$limit':10}
]
```

```python
for i in item_info.aggregate(pipeline):
    print(i)
```

```python
data = [ [i['_id'],i['counts']] for i in item_info.aggregate(pipeline)]
print(data)
```

```python
series =  [{
    'type': 'pie',
    'name': 'Browser share',
    'data': data
}]

charts.plot(series, options={'title': {'text': '某一时间点斗鱼颜值主播的地域分布饼图'}}, show='inline')
```
